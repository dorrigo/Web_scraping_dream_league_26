```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('dota2_dream_league_26.csv')
print(dataset.head())
```

```python
# Mostra os her칩is mais pickados
data_pick = dataset.sort_values(by='Picks', ascending=False).head(5)[['Hero', 'Picks']]

# Cria o gr치fico de barras com os 5 her칩is mais pickados
sns.barplot(x='Picks', y='Hero', data=data_pick, palette='viridis', edgecolor='black')
plt.title('Top 5 Her칩is mais Pickados')
for i, (hero, picks) in enumerate(zip(data_pick['Hero'], data_pick['Picks'])):
    plt.text(picks + 1, i, str(picks), va='center')
plt.figure(figsize=(10, 6))
plt.show()
```

```python
# Define dataset para os 5 her칩is mais banidos
data_ban = dataset.sort_values(by='Bans', ascending=False).head(5)[['Hero', 'Bans']]

# Cria o gr치fico de barras para os 5 her칩is mais banidos
sns.barplot(x='Bans', y='Hero', data=data_ban, palette='rocket', edgecolor='black')
plt.title('Top 5 Her칩is mais Banidos')
for i, (hero, bans) in enumerate(zip(data_ban['Hero'], data_ban['Bans'])):
    plt.text(bans + 1, i, str(bans), va='center')
plt.figure(figsize=(10, 6))
plt.show()
```

```python
# Filtra her칩is com pelo menos 10 picks e ordena pelo maior win rate
data_winrate = dataset[dataset['Picks'] >= 10][['Hero', 'Win Rate']].copy()
# Converte 'Win Rate' de string percentual para float
data_winrate['Win Rate'] = data_winrate['Win Rate'].str.rstrip('%').astype(float) / 100
data_winrate = data_winrate.sort_values(by='Win Rate', ascending=False).head(5)

# Cria o gr치fico de barras para os 5 her칩is com maior win rate
sns.barplot(x='Win Rate', y='Hero', data=data_winrate, palette='mako', edgecolor='black')
plt.title('Top 5 Her칩is com Maior Win Rate (Picks >= 10)')
for i, (hero, win_rate) in enumerate(zip(data_winrate['Hero'], data_winrate['Win Rate'])):
    plt.text(win_rate + 0.01, i, f"{win_rate:.2%}", va='center')
plt.figure(figsize=(10, 6))
plt.show()
```

```python
# Seleciona os 5 her칩is com pior win rate (pelo menos 10 picks)
data_worst_winrate = dataset[dataset['Picks'] >= 10][['Hero', 'Win Rate']].copy()
data_worst_winrate['Win Rate'] = data_worst_winrate['Win Rate'].str.rstrip('%').astype(float) / 100
data_worst_winrate = data_worst_winrate.sort_values(by='Win Rate').head(5)
# Cria o gr치fico de barras para os 5 her칩is com pior win rate
sns.barplot(x='Win Rate', y='Hero', data=data_worst_winrate, palette='coolwarm', edgecolor='black')
plt.title('Top 5 Her칩is com Pior Win Rate (Picks >= 10)')
for i, (hero, win_rate) in enumerate(zip(data_worst_winrate['Hero'], data_worst_winrate['Win Rate'])):
    plt.text(win_rate + 0.01, i, f"{win_rate:.2%}", va='center')
plt.figure(figsize=(10, 6))
plt.show()
```
